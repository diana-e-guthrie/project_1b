```python
happy19=pd.DataFrame(pd.read_csv('happiness_2019.csv'))
happy19.head()
```

```python
data19=pd.DataFrame(data)
data19=data19[['name','gini','population','area']]
data19=data19.rename(columns={'name':'Country'})
data19.head()
```

```python
happy_pop=pd.merge(happy19, data19, on='Country', how='inner')
happy_pop
```

```python
happy_pop=happy_pop.dropna(how='any')

len(happy_pop)
```

```python
def lregress(x_values, y_values):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.title(input('Title of figure:'))
    plt.xlabel(input('X-label:'))
    plt.ylabel(input('Y-label:'))
    x=float(input('x-value for location of line equation'))
    y=float(input('y-value for location of line equation'))
    plt.annotate(line_eq,(x,y),fontsize=15,color="red")
    plt.grid()
    plt.show()
    print(f"R-value: {rvalue}, P-value: {pvalue}")
```

```python
happiness=happy_pop['Happiness score']
gini=happy_pop['gini']

x_values=happiness
y_values=gini
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.title('Happiness vs GINI index (measure of economic inequality)')
plt.xlabel('Happiness Score')
plt.ylabel('GINI Index')
plt.annotate(line_eq,(5,60),fontsize=15,color="red")
plt.grid()
plt.show()
print(f"R-value: {rvalue}, P-value: {pvalue}")
```
